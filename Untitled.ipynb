# Supervised Algorithms

We're using two supervised classification algorithms: K-Nearest Neighbors and Naive Bayes.<br>
<br>
**K-Nearest Neighbors**: Classify based on the majority class "near" the data point<br>
**Naive Bayes**: Predicts the class that has the highest probability given the input features. See Bayes Theorem.

```python
mall.head()
```

### KNN prep

KNN requires numerical input data and it requires scaling of the numerical columns. I'll use one-hot encoding for gender, 

```python
# One hot encoding for gender column
mall_knn = pd.get_dummies(mall, columns = ['gender'])

mall_knn['gender_Female'], mall_knn['gender_Male'] =  mall_knn['gender_Female'].astype(int), mall_knn['gender_Male'].astype(int)
```

```python
mall_knn
```

Age and Income are normally distributed enough for a StandardScalar.

```python
scaler = StandardScaler()
mall_knn[['age', 'income']] = scaler.fit_transform(mall_knn[['age', 'income']])
```

```python
mall_knn[['age', 'income']].head()
```

And finally, we'll map the spending_level classes to integers.<br>
0 = low<br>
1 = medium<br>
2 = high<br>

```python
spend_map = {
    'low': 0,
    'medium': 1,
    'high': 2
}

mall_knn['spending_level'] = mall_knn['spending_level'].map(spend_map)
print(mall_knn['spending_level'].value_counts())
```

KNN time!

```python
# Our training and test sets
X = mall_knn[['gender_Male', 'gender_Female', 'age', 'income']].copy()
y = np.array(mall_knn[['spending_level']].copy()).reshape(len(mall_knn),)

X_train, X_test, y_train, y_test = train_test_split(
    X, 
    y, 
    test_size=0.2, 
    random_state = 13,
    stratify = y
)
```

```python
model = KNeighborsClassifier(n_neighbors=3)
model.fit(X_train, y_train)


y_pred = model.predict(X_test)

classes = spend_map.keys()

cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot = True, xticklabels =classes, yticklabels = classes, cmap = 'Blues')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix')
plt.show()
print(classification_report(y_test, y_pred))

print("Accuracy:", accuracy_score(y_test, y_pred))
```

Eh. It's okay. I'll use cross_validation to check how well the model performs on different subsets.<br>
<br>
Note: I came back later and added the StratiedKFold to make sure my dataset was split into sets with balanced target classes.

```python
cv = StratifiedKFold(n_splits = 5, shuffle = True, random_state = 13)


model = KNeighborsClassifier(n_neighbors=3)
scores = cross_val_score(model, X, y, cv=cv, scoring = 'accuracy' )

print("Cross-validated scores:", scores)
print(f"Mean accuracy: {scores.mean():.2f}")
```

It's okay. Let's try a GridSearch to see if improving n will help. 

```python

param_grid = {
    'n_neighbors': list(range(1,25))
}

grid = GridSearchCV(
    estimator = KNeighborsClassifier(), 
    param_grid = param_grid,
    cv = cv,
    scoring = 'accuracy',
    n_jobs = -1,
    verbose = 2
)

grid.fit(X_train, y_train)

print("Best parameters found:", grid.best_params_)
print(f"Best CV Score: {grid.best_score_:.4f}")

accuracy = grid.score(X_test, y_test)
print(f"Accuracy: {accuracy:.4f}")
```

```python
pred = grid.predict(X_test)

cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot = True, xticklabels =classes, yticklabels = classes, cmap = 'Blues')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix')
plt.show()

print(classification_report(y_test, pred))
```

### BernoulliNB

```python
mall.head()
```

```python
mall_knn.head()
```

```python
mall_bnb = pd.concat([mall[['age', 'income', 'spending_level']], mall_knn[['gender_Female', 'gender_Male']]], axis = 1)
```

```python
mall_bnb
```

To prepare the dataset for BernoulliNB I'm going to convert the income and age columns into binary. I'll separate income by the median income. Anything above or equal will be 1, else 0. I'll do the same for age.

```python
age_median = mall.age.median()
income_median = mall.income.median()

mall_bnb.loc[mall['age'] < age_median ,'age_binary'] = 0
mall_bnb.loc[mall['age'] >= age_median ,'age_binary'] = 1

mall_bnb.loc[mall['income'] < income_median ,'income_binary'] = 0
mall_bnb.loc[mall['income'] >= income_median ,'income_binary'] = 1
```

```python
mall_bnb.head()
```

```python
X = mall_bnb[['gender_Male', 'gender_Female', 'age_binary', 'income_binary']].copy()
y = mall_knn['spending_level']

X_train, X_test, y_train, y_test = train_test_split(
    X, 
    y, 
    test_size=0.2, 
    random_state = 13,
    stratify = y
)


model = BernoulliNB()

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("Test set accuracy:", accuracy)
```

```python
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot = True, xticklabels =classes, yticklabels = classes, cmap = 'Blues')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix')
plt.show()

print(classification_report(y_test, y_pred))
```

Haha, that's pretty bad. It's almost as bad as picking randomly. The only thing that saves it is that the model has good recall for high (predicts positive cases correctly). Buuuut it also has really low precision for high; out of the 22 times it predicted high, only 11 of them were correct. I'll try again with a gridsearch.

```python
params = param_grid = {
    'alpha': [0.1, 0.5, 1.0, 2.0, 5.0],        
    'fit_prior': [True, False]    
}

grid = GridSearchCV(
    estimator = BernoulliNB(), 
    param_grid = params,
    cv = cv,
    scoring = 'accuracy',
    n_jobs = -1,
    verbose = 2
)

grid.fit(X_train, y_train)

print("Best parameters found:", grid.best_params_)
print(f"Best CV Score: {grid.best_score_:.4f}")

accuracy = grid.score(X_test, y_test)
print(f"Accuracy: {accuracy:.4f}")
```

```python
y_pred = grid.predict(X_test)

cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot = True, xticklabels =classes, yticklabels = classes, cmap = 'Blues')

plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix')
plt.show()

print(classification_report(y_test, y_pred))
```

No difference. One last try before moving on. Instead of manually selecting the threshold value to binarize age and income, I'll use gridsearch to try multiple thresholds.<br>
First, I'll normalize the age and income columns with a min-max scalar to make their ranges match. I'll apply the scalar and model in a pipeline this time.

```python
X = pd.concat([mall[['age', 'income']], mall_knn[['gender_Female', 'gender_Male']]], axis = 1)
y = mall_knn.spending_level

X_train, X_test, y_train, y_test = train_test_split(
    X, 
    y, 
    test_size=0.2, 
    random_state = 13,
    stratify = y
)
```

```python
pipeline = Pipeline([
    ('scalar', MinMaxScaler()),
    ('bnb', BernoulliNB())
])

params = param_grid = {
    'bnb__alpha': [0.1, 0.5, 1.0, 2.0, 5.0],        
    'bnb__fit_prior': [True, False],
    'bnb__binarize' : [i/10 for i in range(0,11)]
}

grid = GridSearchCV(
    pipeline,
    param_grid = params,
    cv = cv,
    scoring = 'accuracy',
    n_jobs = -1,
    verbose = 2
)

grid.fit(X_train, y_train)

print("Best parameters found:", grid.best_params_)
print(f"Best CV Score: {grid.best_score_:.4f}")

accuracy = grid.score(X_test, y_test)
print(f"Accuracy: {accuracy:.4f}")
```

```python
y_pred = grid.predict(X_test)

cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot = True, xticklabels =classes, yticklabels = classes, cmap = 'Blues')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix')
plt.show()

print(classification_report(y_test, y_pred))
```

Yeah, this ain't it. A simple binary does not adequately capture the relationships between income or age and spending level. I'm going to move on to unsupervised algorithms now.

# Unsupervised Algorithms

## K-Means

```python
mall_kmeans = pd.concat([mall[['age','income', 'spending_score', 'gender']], mall_knn[['gender_Male', 'gender_Female',]]], axis = 1)

scaler = StandardScaler()
mall_kmeans[['age_std', 'income_std','spending_std']] = scaler.fit_transform(mall_kmeans[['age', 'income', 'spending_score']])

X = mall_kmeans[['age_std', 'income_std','spending_std', 'gender_Male', 'gender_Female']]
X.head()
```

```python
model = KMeans(n_clusters = 3, random_state=13)

mall_kmeans['cluster'] = model.fit_predict(X)
```

```python
#initialize kmeans parameters
kmeans_kwargs = {
"init": "random",
"n_init": 10
}

#create list to hold SSE values for each k
sse = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, **kmeans_kwargs)
    kmeans.fit(X)
    sse.append(kmeans.inertia_)

#visualize results
plt.plot(range(1, 11), sse)
plt.xticks(range(1, 11))
plt.xlabel("Number of Clusters")
plt.ylabel("SSE")
plt.show()
```

## Agglomerative Clustering

Agglomerative Clustering is a type of Hierarchical Clustering, which is a type of clustering that links a dataset by a measure of similarity. Agglomerative just means that the algorithm starts with a cluster for each datapoint and starts linking them. The 2nd type of Hierarchical Clustering called Partial Clustering, starts with one cluster and breaks it into smaller and smaller clusters based on the similarity between points. This will do exactly the same as kmeans but in a different way.

I'm going to use the mall_kmeans dataframe again, and the same X I built in KMeans.

```python
X.head()
```

```python
model = AgglomerativeClustering(n_clusters=3) 

mall_kmeans["agglo_cluster"] = model.fit_predict(X)
```

```python
sns.set(style="white")

fig, ax = plt.subplots(figsize=(12, 8)) 
sns.scatterplot(
    data=mall_kmeans,
    x="income",
    y="spending_score",
    hue="agglo_cluster",
    style="gender",
    size="age",
    sizes=(40, 200),
    palette="Set2",
    alpha=0.8,
    edgecolor="black"
)

ax.grid(False)

plt.title("Customer Segments (Agglo Clusters)", fontsize=14, weight = 'bold')
plt.xlabel("Income")
plt.ylabel("Spending Score")
plt.legend(bbox_to_anchor=(1, 1), loc='upper left',) 

sns.despine(ax=ax)

plt.show()
```

```python
mall_kmeans.agglo_cluster.value_counts()
```

This actually looks pretty close to the first K-Means cluster. (After further research this makes sense: the sklearn function by default is using Euclidean distance to calculate similarity and is minimizing the variance between datapoints in each cluster, same as KMeans) I'll continue tuning to see if any more patterns emerge. <br>
